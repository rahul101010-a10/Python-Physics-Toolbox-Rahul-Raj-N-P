Task 1 – Plasma Temperature Classifier (Loops + Conditionals + Functions)
Write a function classify_plasma_temp(temp) that takes temperature (float, in Kelvin) and returns: • "Cold plasma" (< 10^4 K) • "Warm plasma" (10^4 – 10^6 K) • "Hot plasma" (> 10^6 K) Use it in a loop: ask user for 5 temperatures (handle invalid input with try-except), classify each, and store results in a list of tuples (temp, category). Print nicely formatted table. Add 1–2 sentences: why temperature classification matters in plasma physics experiments.

```python
import numpy as np

def classify_plasma_temp(temp):
    if temp < 1e4:
        return "Cold plasma"
    elif 1e4 <= temp <= 1e6:
        return "Warm plasma"
    else:
        return "Hot plasma"

temps = []
results = []

for i in range(5):
    try:
        t = float(input(f"Enter temperature {i+1} in Kelvin: "))
        temps.append(t)
        results.append(classify_plasma_temp(t))
    except ValueError:
        print("Invalid input. Please enter numeric value.")

temps_array = np.array(temps)

print("\n--- Plasma Classification Table ---")
for t, r in zip(temps_array, results):
    print(f"{t:.2e} K  -->  {r}")
```

# Task 2 – Material Properties Lookup (Dictionaries + Lists + Strings)
Create a dictionary with 4–5 materials from materials science (e.g. Aluminum, Copper, Silicon, Graphene). Each value = another dict: {'density_g_cm3': 2.7, 'conductivity': 3.5e7, 'youngs_modulus_GPa': 70}
Write functions to:
•	Add a new material
•	Find material with highest conductivity
•	Return list of densities as floats (use list comprehension + string cleaning if needed)
Show example usage + short comment on how dicts help organize experimental data.

```python
import numpy as np

materials = {
    "Aluminum": {"density_g_cm3": 2.7, "conductivity": 3.5e7, "youngs_modulus_GPa": 70},
    "Copper": {"density_g_cm3": 8.96, "conductivity": 5.9e7, "youngs_modulus_GPa": 110},
    "Silicon": {"density_g_cm3": 2.33, "conductivity": 1e-3, "youngs_modulus_GPa": 130},
    "Graphene": {"density_g_cm3": 2.2, "conductivity": 1e8, "youngs_modulus_GPa": 1000}
}

def highest_conductivity(materials):
    return max(materials, key=lambda x: materials[x]["conductivity"])

def get_densities(materials):
    return np.array([materials[m]["density_g_cm3"] for m in materials])

print("Material with highest conductivity:", highest_conductivity(materials))
print("Densities array:", get_densities(materials))
```

# Task 3 – Simple Particle Energy Calculator (Functions + OOP basics)
Define a class Particle with attributes: mass_kg, velocity_m_s, name (str)
Add method kinetic_energy() → return ½mv² (use float)
Create 3 particle objects (e.g. electron, proton, alpha particle with realistic masses/velocities). Loop over them → print name + energy in eV (convert using 1 eV ≈ 1.6e-19 J).
Bonus (extra 2 marks if time): add charge attribute and simple potential_energy(q, V) method.


```python
import numpy as np

eV_conversion = 1.6e-19

class Particle:
    def __init__(self, name, mass_kg, velocity_m_s):
        self.name = name
        self.mass = mass_kg
        self.velocity = velocity_m_s
        
    def kinetic_energy(self):
        return 0.5 * self.mass * self.velocity**2
    
    def kinetic_energy_eV(self):
        return self.kinetic_energy() / eV_conversion

electron = Particle("Electron", 9.11e-31, 2e6)
proton = Particle("Proton", 1.67e-27, 1e5)
alpha = Particle("Alpha Particle", 6.64e-27, 5e5)

particles = [electron, proton, alpha]

energies = []

for p in particles:
    energy = p.kinetic_energy_eV()
    energies.append(energy)
    print(f"{p.name} Energy = {energy:.2e} eV")

energies_array = np.array(energies)
```
